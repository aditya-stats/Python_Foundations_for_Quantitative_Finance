Pandas Introduction 

This notebook covers series, dataframes, and read CSV

```python
import pandas as pd
```

### Series

```python
a = [1, "o", 2, "p"]

var = pd.Series(a)

print(var)
```

```python
#Labels
var = pd.Series(a, index = ["z", "x", "c", "v"])
print(var)
print(var["x"])
```

### DataFrames

```python
data = {
  "Sales": [20, 30, 13],
  "Duration(in hrs.)": [30, 40, 25]
}
dframe = pd.DataFrame(data, index = [1,2,3])

print(dframe)
```

```python
#to return a row
print(dframe.loc[[2,1]])
```
