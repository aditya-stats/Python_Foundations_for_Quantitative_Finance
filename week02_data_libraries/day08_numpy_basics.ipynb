### Accessing / Slicing / Changing specific rows, columns

```python
a = np.array([[2,4,6,8,10],[1,3,5,7,9]])
print(a)
```

To get a specific element

```python
a[0, 2]
```

```python
#Specific column
a[0:,2]
```

```python
#Specific row
a[1,:]
```

```python
#[start:stop:step]
a[1,0:5:2]
```

```python
b =np.array([[[2, 4],[6, 8]],[[1, 3],[5, 7]]])
print(b)
```

```python
#To get 8
b[0,1,1]
```

Replace

```python
b[1,0,:] = [11,33]
```

Repating an array

```python
ar1 = np.array([[5, 50, 500]])
rep1 = np.repeat(ar1, 4, axis=0)
print(rep1)
```

```python
ar1 = np.array([[[[5, 50, 500]]]])
rep1 = np.repeat(ar1, 4, axis=0)
print(rep1)
```
