```python
#Plotting with keyword strings
data = { 'a' : np.arange(30),
         'b' : np.random.randint(0, 30, 30),
         'c' : np.random.randn(30) }
data["d"] = data['a'] + 10*data['c']
data['c'] = np.abs(data['c']) * 10

plt.scatter('a', 'd', c='b', s='c', data=data)
plt.xlabel('entry a')
plt.ylabel('entry d')
plt.show()
```

```python
#Texts in plots
mu, sigma = 100, 13.5
x = mu + sigma * np.random.randn(1000000)
x
```

```python
#n(array containing values of hist.), bin(array of bin edges), patches(collection of rectangle bars )
n, bins, patches = plt.hist(x, 50, density=True, facecolor='g', alpha=0.75)
plt.show()
print(n)
print(bins)
print(patches)
```

### Histograms

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import colors
from matplotlib.ticker import PercentFormatter
```

```python
rng = np.random.default_rng(2)
```

```python
n_points = 2000
n_bins = 20

dist1 = rng.standard_normal(n_points)
dist2 = 0.4 * rng.uniform(-4, 4, n_points)
dist3 = rng.binomial(n_points, 0.0003)
fig, axs = plt.subplots(1, 3, sharey=True, tight_layout=True)
print(fig, axs)

axs[0].hist(dist1, bins=n_bins)
axs[1].hist(dist2, bins=23)
axs[2].hist(dist3, bins=n_bins)

plt.show()
```
