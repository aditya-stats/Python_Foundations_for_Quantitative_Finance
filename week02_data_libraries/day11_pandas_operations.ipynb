### Filtering

```python
df = pd.read_csv(r"C:\Users\adity\Downloads\annual-enterprise-survey-2024-financial-year-provisional.csv")
print(df)
```

```python
#accessing a book object
df.Value
```

```python
df["Value"]
```

```python
#to get a specific value
df["Value"][3]
```

```python
#to select 1st row of the dataframe
df.iloc[0]
```

```python
#to get the 1st column with iloc - index basedselection
df.iloc[:, 0]
```

```python
#to get specific rows and columns
df.iloc[[0,5,10,-55555],[-1,5]]
```

```python
#loc - label based selection
df.loc[2500:2505, ['Variable_code', 'Industry_code_ANZSIC06', 'Value']]
```

```python
#Manipulating index
df.set_index("Year")
```

```python
#Boolean filtering
df.Variable_code == "H05"
```

```python
#to select relevant data
df.loc[df.Variable_code == "H05"]
```

```python
#& - 2 conditions together
df.loc[(df.Variable_code == "H05") & (df.Industry_aggregation_NZSIOC == "Level 3")]
```

```python
# | - either of the conditions
df.loc[(df.Variable_code == "H20") | (df.Industry_aggregation_NZSIOC == "Level 4")]
```

```python
# isin
df.loc[df.Variable_code.isin(["H05", "H10"])]
```

```python
#isnull
df.loc[df.Industry_code_ANZSIC06.isnull()]  #no empty values
```

```python
#assigning data
df['H05'] = "M05"
df['H05']
```

```python
df['index_backwards'] = range(len(df), 0, -1)
df['index_backwards']
```

### Sorting

```python
indus_counted = df.groupby('Industry_code_ANZSIC06').Industry_code_ANZSIC06.count()
indus_counted.sort_values()
```

```python
indus_counted.sort_values(ascending = False)
```

```python
indus_counted.sort_index()
```
