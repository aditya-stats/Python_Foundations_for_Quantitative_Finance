This notebook covers probability simulations

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Coin Simulations

```python
#Single coin toss
toss = np.random.choice([0,1])
print(toss)
```

```python
#Many coin tosses
Tosses = np.random.choice([0,1], size=1000)
print("First 25 tosses:")
print(Tosses[:25])
```

```python
heads = np.sum(Tosses)
print("Number of heads:", heads)
```

```python
prob_heads = np.mean(Tosses)

print("Empirical probability of heads:", prob_heads)
```

```python
plt.hist(Tosses, bins=2)

plt.title("Distribution of Coin Toss Outcomes")
plt.xlabel("Outcome (0 = Tail, 1 = Head)")
plt.ylabel("Frequency")

plt.show()
```

```python
#Law of Large Numbers Simulation
n = 100000
tosses = np.random.choice([0,1], size=n)

running_mean = np.cumsum(tosses) / np.arange(1, n+1)

plt.plot(running_mean)

plt.axhline(0.5)

plt.title("Law of Large Numbers")
plt.xlabel("Number of Trials")
plt.ylabel("Sample Mean")

plt.show()
```

### Dice Simulations

```python
dice = np.random.randint(1,7)

print("Dice roll:", dice)
```

```python
#Many dice rolls
rolls = np.random.randint(1,7, size=10000)

print("First 999 rolls:")
print(rolls[:999])
```

```python
mean_roll = np.mean(rolls)

print("Empirical mean:", mean_roll)
```

Theoretical mean: 3.5

```python
#Plotting dice distribution
plt.hist(rolls, bins=6)

plt.title("Dice Roll Distribution")
plt.xlabel("Dice Outcome")
plt.ylabel("Frequency")

plt.show()
```

```python
#2 dice simulation
dice1 = np.random.randint(1,7,100000)
dice2 = np.random.randint(1,7,100000)
sum_dice = dice1 + dice2
prob_7 = np.sum(sum_dice == 7) / len(sum_dice)
print("Probability that sum = 7 is", prob_7)
```

```python
#plot
plt.hist(sum_dice, bins=11)
plt.title("Distribution of Sum of Two Dice")
plt.xlabel("Sum")
plt.ylabel("Frequency")
plt.show()
```

# Uniform Distribution

```python
x = np.random.uniform(0,1,100000)

mean_x = np.mean(x)
value = np.mean(x**2)

print("Estimated E[X]:", mean_x)
print("Estimated E[X^2]:", value)

Var = value - mean_x**2
print("Variance:", Var)
```

# Binomial Distribution

```python
binom = np.random.binomial(10, 0.5, size=10000)

print("First 20 values:")
print(binom[:20])
mean_binom = np.mean(binom)
print("Empirical mean:", mean_binom) #Expected empirical mean = 5
```

```python
#Plot-
plt.hist(binom, bins=11)
plt.title("Binomial Distribution Simulation")
plt.xlabel("Number of Successes")
plt.ylabel("Frequency")

plt.show()
```
