Descriptive Statistics

This notebook covers mean, varience, skewness, kurtosis

```python
# ensure required libraries are installed
%pip install matplotlib pandas numpy scipy

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import skew, kurtosis as scipy_kurtosis

from stats_functions import mean, variance, std_dev, skewness, kurtosis
```

```python
#Generate simulated data
np.random.seed(42)

# Heavy-tailed distribution
returns = np.random.standard_t(df=4, size=1000)

returns = pd.Series(returns)
```

```python
#Using Pandas built-in function
returns.describe()
```

```python
#Use the functions to compute statistics
print("Mean:", mean(returns))
print("Sample Variance:", variance(returns, ddof=1))
print("Std Dev:", std_dev(returns, ddof=1))
print("Skewness:", skewness(returns))
print("Excess Kurtosis:", kurtosis(returns))
```

```python
#Comparision with library functions
print("NumPy Mean:", np.mean(returns))
print("NumPy Sample Variance:", np.var(returns, ddof=1))
print("SciPy Skewness:", skew(returns))
print("SciPy Excess Kurtosis:", scipy_kurtosis(returns))
```

```python
#Visualize the distribution
plt.figure(figsize=(8,5))
plt.hist(returns, bins=40)
plt.title("Distribution of Simulated Returns")
plt.xlabel("Returns")
plt.ylabel("Frequency")
plt.show()
```

Conclusion: 
The simulated returns exhibit heavy tails, as indicated by the high excess kurtosis. The skewness is close to zero, suggesting a symmetric distribution.
The value found using library functions are very close to mannual functions used.
